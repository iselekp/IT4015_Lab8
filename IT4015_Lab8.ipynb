## Task 3: Data Cleaning

### Filling missing values

I fill in the missing values with a default value that is representative of the overall distribution - the median. I don't use interpolate, ffill or bfill because the consecutive entries are not necessarily correlated directly with each other. That is, consecutive entries may have different categorical values and the order/index is irrelevant. 

```python
median_daysOpen = df["daysOpen"].median(skipna=True)
```

```python
median_daysOpen
```

```python
df = df.fillna(median_daysOpen)
```

### Correcting inconsistencies

I replace the inconsistent data entries with the most similar standard entries.

```python
df = df.replace("Hardware!", "Hardware")
df = df.replace("Acc3ss/Login", "Access/Login")
df = df.replace("SyStemsEror", "Systems")
df = df.replace("softwaRE+", "Software")
```
